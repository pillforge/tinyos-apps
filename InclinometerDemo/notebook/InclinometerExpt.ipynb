# Inclinometer Experiment

The Final tuning parameters were
$$K_p = 10.0$$ 
$$K_d = 20.0$$

The PWM was scaled in order for the motor to generate enough torque and move for all error values. The scaling factor($K$) was $0.4$ and the scaling offset($O$) was $140$. The resulting expression for an 8 bit PWM output was:
$$ PWM = x*K + O $$

where $x$ is the result of the standard PID expression

This results in the PWM output range of $140-200$

```python
data = loadtxt('../data1.csv', skiprows=1)
datavalid = data[350:500]
```

```python
ts, angle = datavalid[:,[0,2]].T
```

```python
figure(figsize=(14,8))
plot(ts-ts[0], angle)
xlabel("Time (s)")
ylabel("Angle (Degrees)")
grid(True)
```
